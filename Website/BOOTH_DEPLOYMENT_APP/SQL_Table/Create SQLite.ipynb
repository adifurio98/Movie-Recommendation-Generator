```python
from sqlalchemy import create_engine, inspect

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = "titanic.csv"
df = pd.read_csv(data)
df.head()
```

```python
database_path = "titanic.sqlite"
conn_string = f"sqlite:///{database_path}"

# Create an engine that can talk to the database
engine = create_engine(conn_string)
```

```python
sql = """CREATE TABLE titanic (
    passengerid int,
    survived bool,
    pclass int,
    name varchar(500),
    sex varchar(10),
    age int,
    parch int,
    ticket varchar(50),
    fare double,
    cabin varchar(10),
    embarked varchar(5)
    )"""
```

```python
engine.execute(sql)
```

```python
# get all tables
inspector_gadget = inspect(engine)
tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```

```python
df.to_sql("titanic", con=engine, method="multi", index=False, if_exists="replace")
```

```python
# Make a connection to the SQL database
conn = engine.connect()

query = """
        SELECT
            *
        FROM
            titanic;
        """

df2 = pd.read_sql(query, conn)
df2.head()
```
