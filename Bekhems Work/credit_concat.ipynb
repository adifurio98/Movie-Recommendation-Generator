```python
def order_remove(x):
    new_list = []
    for dic in x:
        new_dic = {} 
        for d in dic.keys():
            if d not in ['order', 'profile_path']:
                new_dic[d] = dic[d]             
        new_list.append(new_dic)
    return new_list

```
